## Importing data into Neo4j:

First of all we have to ensure the integrity of the schema, so we create the constraint of uniqueness of character names.

CREATE CONSTRAINT ON (c:Character) ASSERT c.name IS UNIQUE;

Now let's load data: 

LOAD CSV WITH HEADERS FROM "file:/stormofswords.csv" AS row
MERGE (src:Character {name: row.Source})
MERGE (tgt:Character {name: row.Target})
MERGE (src)-[r:INTERACTS]->(tgt)
ON CREATE SET r.weight = toInteger(row.Weight)
return src, tgt

Example query:

MATCH (Jon:Character {name: "Jon"})<-[:INTERACTS]-(characters) RETURN Jon,characters

## Exploring the graph in Neo4j browser:

Count the number of nodes:

MATCH (character:Character)RETURN count(*)

returns 107 nodes.

Summary of interaction between characters: 

MATCH (character:Character)-[r:INTERACTS]->() count(*) AS num RETURN min(num),max(num),avg(num)

```python
Image(filename='interSum.png')
```

#### Shortest path: (Between Catelyn Stark and Kahl Drogo)

MATCH (catelyn:Character {name: "Catelyn"}), (drogo:Character {name: "Drogo"})
MATCH p=shortestPath((catelyn)-[INTERACTS*]-(drogo))
RETURN p

```python
Image(filename='catdro.png')
```

#### all short paths:

MATCH (catelyn:Character {name: "Catelyn"}), (drogo:Character {name: "Drogo"})
MATCH p=allshortestPath((catelyn)-[INTERACTS*]-(drogo))
RETURN p

```python
Image(filename='allspa.png')
```

### The diameter of your graph:

MATCH (a:Character), (b:Character) WHERE id(a) > id(b)
MATCH p=shortestPath((a)-[:INTERACTS*]-(b))
RETURN length(p) AS len,p
ORDER BY len DESC
LIMIT 1

```python
Image(filename='diameter.png')
```

The degrees for the nodes:
    
MATCH (a:Character)
RETURN a.name,size((a)-[:INTERACTS]->()) as out_degree,size((a)<-[:INTERACTS]-()) as in_degree
ORDER BY out_degree DESC,in_degree DESC

```python
Image(filename='degrees.png')
```

#### Weighted degree:

MATCH (a:Character)-[r:INTERACTS]-() RETURN a.name,sum(r.weight) as weighted_degree ORDER BY weighted_degree DESC

```python
Image(filename='wdegree.png')
```
